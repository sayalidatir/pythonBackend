# Download file

```python
import requests
url=("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=")
res=requests.get(url)
print(res.status_code)

with open("index.html","w") as f:
    f.write(res.text)
```

# request module methods

```python
print(res.status_code)
print(res.headers) #metadata
print(res.content)
print(res.json())
print(res.encoding)
```

# Http methods

```python
1.GET requests
```

```python
payload={'key1':'value1','key2':'value2'}
r=requests.get("https://httpbin.org/get",params=payload)  #params used in get
print(r.json()) #parse json data into python dictionary

print(type(r.json()))

#https://httpbin.org/get?key1=value1&key2=value2 to send data through url
```

```python
2.Post requests
```

```python
r=requests.post("https://httpbin.org/post?a=b",data={'sayali':'value1'}) #data and json used here
print(r.text)
```

```python
3.PUT :create new something
```

```python
res=requests.put("https://httpbin.org/put",data={'a':1,'b':3})
print(res.text)
```

# downloding images

```python
from PIL import Image
from io import BytesIO

r=requests.get("https://images.pexels.com/photos/268533/pexels-photo-268533.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500")
i = Image.open(BytesIO(r.content))        
fp=open("img.jpg","wb") #wb binary form
i.save(fp)
fp.close()
```
